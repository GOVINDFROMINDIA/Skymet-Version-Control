```python
import pandas as pd
data_path = '/content/drive/MyDrive/skymet/data.csv'
data = pd.read_csv(data_path)
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
data.set_index('Timestamp', inplace=True)

plt.figure(figsize=(12, 6))
sns.lineplot(x=data.index, y='Ambient Temperature(Deg.C)', data=data)
plt.title('Ambient Temperature Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Ambient Temperature (Deg.C)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Ambient Temperature(Deg.C)', y='Ambient Humidity(%)', data=data)
plt.title('Scatter Plot: Ambient Temperature vs. Humidity')
plt.xlabel('Ambient Temperature (Deg.C)')
plt.ylabel('Ambient Humidity (%)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Precipitation(mm/hour)'], bins=20, kde=True)
plt.title('Precipitation Distribution')
plt.xlabel('Precipitation (mm/hour)')
plt.ylabel('Frequency')
plt.show()
```
